## Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, layout='constrained')

ax = axs[0][0]
ax.plot(np.arange(0, 1e6, 1000))
ax.set_title('Title0 0')
ax.set_ylabel('YLabel0 0')

ax = axs[0][1]
ax.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
ax.set_title('Title0 1')
ax.xaxis.tick_top()
ax.tick_params(axis='x', rotation=55)


for i in range(2):
    ax = axs[1][i]
    ax.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
    ax.set_ylabel('YLabel1 %d' % i)
    ax.set_xlabel('XLabel1 %d' % i)
    if i == 0:
        ax.tick_params(axis='x', rotation=55)

fig.align_labels()  # same as fig.align_xlabels(); fig.align_ylabels()
fig.align_titles()

plt.show()
```

## 目盛り、ラベル

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ticks = ax.set_xticks([0,250, 500, 750, 1000])
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=30, fontsize="small")
ax.set_title("My graph")
ax.set_xlabel("Stages")
ax.plot(np.random.randn(1000).cumsum(), label="one")
ax.plot(np.random.randn(1000).cumsum(), label="two")
ax.plot(np.random.randn(1000).cumsum(), label="three")
ax.legend(loc="best")
```

## 散布図

```python
import pandas as pd
import numpy as np
```

```python
macro = pd.read_csv("../pydata-book/examples/macrodata.csv")
```

```python
macro
```

```python
data = macro[["cpi", "m1", "tbilrate", "unemp"]]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.regplot(trans_data, x="m1", y="unemp")
plt.title("Changes in log {} versus log {}".format("m1", "unemp"))
```
